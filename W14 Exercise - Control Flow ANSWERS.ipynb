## Exercise 4 — for loop + continue

You are processing a log of page views from a web analytics export.  
Some rows have `None` as the session duration (the user closed the browser immediately).

Loop over the data and:
- Skip rows where `duration_sec` is `None` using `continue`
- Count valid sessions and sum the total duration
- After the loop, print the number of valid sessions, total duration, and average duration

```python
page_views = [
    {"page": "/home",     "duration_sec": 45},
    {"page": "/products", "duration_sec": None},
    {"page": "/about",    "duration_sec": 120},
    {"page": "/home",     "duration_sec": 30},
    {"page": "/contact",  "duration_sec": None},
    {"page": "/products", "duration_sec": 95},
    {"page": "/home",     "duration_sec": 60},
]

valid_count = 0
total_duration = 0

for row in page_views:
    if row["duration_sec"] is None:
        continue
    valid_count += 1
    total_duration += row["duration_sec"]

average_duration = total_duration / valid_count

print(f"Valid sessions: {valid_count} | Total duration: {total_duration} seconds | Average duration: {average_duration:.0f} seconds")
```

## Exercise 5 — Aggregate by category (with missing data)

You have a list of products with their category and units sold.  
Some rows have `None` for `units_sold` — the count was not recorded.

1. Find all unique categories (no imports — just a loop and a list)
2. For each category, calculate the total units sold — use `pass` to skip rows where `units_sold` is `None`
3. Also count how many rows were skipped per category
4. Print a summary, one line per category

```python
products = [
    {"name": "Laptop",      "category": "Electronics", "units_sold": 120},
    {"name": "Desk Chair",  "category": "Furniture",   "units_sold": 45},
    {"name": "Headphones",  "category": "Electronics", "units_sold": 200},
    {"name": "Bookshelf",   "category": "Furniture",   "units_sold": 50},
    {"name": "Keyboard",    "category": "Electronics", "units_sold": None},
    {"name": "Desk Lamp",   "category": "Lighting",    "units_sold": None},
    {"name": "Monitor",     "category": "Electronics", "units_sold": 140},
    {"name": "Floor Lamp",  "category": "Lighting",    "units_sold": 60},
    {"name": "Webcam",      "category": "Electronics", "units_sold": 200},
]

# Step 1: find unique categories
categories = []
for p in products:
    if p["category"] not in categories:
        categories.append(p["category"])

# Step 2 & 3: aggregate per category
for category in categories:
    total = 0
    skipped = 0
    for p in products:
        if p["category"] == category:
            if p["units_sold"] is None:
                skipped += 1
                pass  # nothing to add — units_sold is missing
            else:
                total += p["units_sold"]
    print(f"{category:15s} - total units sold: {total:4d}    skipped: {skipped}")
```

## Exercise 6 — String cleaning with regex & ternary expression

You receive a raw export from a sales system. The `revenue` field is a messy string — different formats, currency symbols, dots and commas used inconsistently.

Your job:
1. Use `re.sub()` to strip everything that is not a digit, then convert to `float`
2. Write the cleaned value back into each dict as a new key `"revenue_clean"`
3. Use a ternary to also add a `"tier"` key: `"Large"` if revenue ≥ 50 000, otherwise `"Small"`
4. Print the cleaned raw_sales

```python
import re

raw_sales = [
    {"region": "North",   "revenue": "85,000 kr"},
    {"region": "South",   "revenue": "SEK 42.500"},
    {"region": "East",    "revenue": "120,000kr"},
    {"region": "West",    "revenue": "37 800 SEK"},
    {"region": "Central", "revenue": "SEK 61.200"},
]

for row in raw_sales:
    # Strip everything that is not a digit
    # Note: dots and commas here are thousands separators, not decimals
    # so we remove them along with everything else
    cleaned = re.sub(r'[^\d]', '', row["revenue"])
    row["revenue_clean"] = float(cleaned)
    row["tier"] = "Large" if row["revenue_clean"] >= 50_000 else "Small"

for row in raw_sales:
    print(row)
```

## Extra: exercise 7 — Dimensional enrichment

You have a fact table of transactions (each row has a `product_id`) and a separate dimension table that maps `product_id` to `category` and `brand` — just like a star schema.

1. Loop over the transactions
2. Look up the product's `category` and `brand` from the dimension dict using `.get()`
3. Add them as new keys directly into each transaction dict
4. If a `product_id` is not found in the dimension (unknown product), set both to `"Unknown"`
5. After enriching all rows, aggregate total revenue by category using another dict as accumulator
6. Print the enriched transactions and the category totals

```python
# Dimension table: this is your lookup / map
product_dim = {
    "P01": {"category": "Electronics", "brand": "TechCo"},
    "P02": {"category": "Furniture",   "brand": "NordicHome"},
    "P03": {"category": "Electronics", "brand": "TechCo"},
    "P04": {"category": "Lighting",    "brand": "BrightLife"},
}

# Fact table: transactions
transactions = [
    {"txn_id": 1, "product_id": "P01", "revenue": 12000},
    {"txn_id": 2, "product_id": "P03", "revenue": 8500},
    {"txn_id": 3, "product_id": "P02", "revenue": 4300},
    {"txn_id": 4, "product_id": "P99", "revenue": 600},
    {"txn_id": 5, "product_id": "P04", "revenue": 3100},
    {"txn_id": 6, "product_id": "P01", "revenue": 9400},
    {"txn_id": 7, "product_id": "P02", "revenue": 5200},
]

# Step 1–4: enrich each transaction with category and brand
for txn in transactions:
    dim = product_dim.get(txn["product_id"])  # returns None if not found
    if dim is None:
        txn["category"] = "Unknown"
        txn["brand"] = "Unknown"
    else:
        txn["category"] = dim["category"]
        txn["brand"] = dim["brand"]

# Step 5 (option 1): aggregate revenue by category with manual accumulator

# create a dict
revenue_by_category = {} 

# manual accumulator
for txn in transactions:
    cat = txn["category"]
    if cat not in revenue_by_category:
        revenue_by_category[cat] = 0
    revenue_by_category[cat] += txn["revenue"]
    
'''
# Step 5 (option 2): aggregate revenue by category + print revenue by category

# create a set
categories = {txn["category"] for txn in transactions} 

# use generator expression
for cat in sorted(categories):
    rows = [txn for txn in transactions if txn["category"] == cat]
    total = sum(txn["revenue"] for txn in rows)
    print(f"{cat:15s}  {total:,}")
    
'''

# Step 6: print results
print("--- Enriched transactions ---")
for txn in transactions:
    print(txn)

print("\n--- Revenue by category ---")
for cat, total in revenue_by_category.items():
    print(f"{cat:15s}  {total:,}")
```

## 🎉 Bonus — input()

Write a small guessing game:

1. Set a secret number (e.g. `42`)
2. Use a `while` loop to keep asking the user to guess with `input()`
3. Tell them if their guess is too high, too low, or correct
4. When they guess correctly, print how many attempts it took and break out of the loop

```python
secret = 42
attempts = 0

while True:
    guess = int(input("Guess the number: "))
    attempts += 1
    if guess < secret:
        print("Too low!")
    elif guess > secret:
        print("Too high!")
    else:
        print(f"Correct! You got it in {attempts} attempt(s).")
        break
```
