## Exercise 1 — if / elif / else

You have a student's exam score. Write an `if/elif/else` block that prints a grade label:

| Score | Label |
|-------|-------|
| 90 and above | `"A"` |
| 75 – 89 | `"B"` |
| 60 – 74 | `"C"` |
| Below 60 | `"F"` |

Test your code with a few different values of `score`.

````python
score = 82

def grade_function(x):
    if x >= 90:
        return"A"
    elif x >= 75 and x <= 89:
        return "B"
    elif x >= 60 and x <= 74:
        return "C"
    elif x < 60:
        return "F"

print(grade_function(score))
````

## Exercise 3 — for + break 

You are scanning a list of usernames to find the first one that starts with the letter `"A"`.  
Use a `for` loop and `break` as soon as when you find it.

Print the name and its position (index) in the list.


````python
usernames = ["Björn", "Clara", "aDavid", "Astrid", "Erik", "Alma"]

for i, name in enumerate(usernames):
    if name.startswith("A"): 
        print(f"Found {name} at {i} position")
        break
````

````python
import re

for i, name in enumerate(usernames):
    if re.match(r"a", name, re.IGNORECASE):
        print(f"Found {name} at {i} position")
        break 
````

## Exercise 4 — for loop + continue

You are processing a log of page views from a web analytics export.  
Some rows have `None` as the session duration (the user closed the browser immediately).

Loop over the data and:
- Skip rows where `duration_sec` is `None` using `continue`
- Count valid sessions and sum the total duration
- After the loop, print the number of valid sessions, total duration, and average duration

**Expected output (example format)**

```python
Valid sessions: 5 | Total duration: 350 seconds  | Average duration: 70 seconds
```

````python
page_views = [
    {"page": "/home",     "duration_sec": 45},
    {"page": "/products", "duration_sec": None},
    {"page": "/about",    "duration_sec": 120},
    {"page": "/home",     "duration_sec": 30},
    {"page": "/contact",  "duration_sec": None},
    {"page": "/products", "duration_sec": 95},
    {"page": "/home",     "duration_sec": 60},
]

valid_count = 0
Total_duration = 0

for page in page_views:
       if page["duration_sec"] is None:
           continue
       Total_duration += page["duration_sec"]
       valid_count += 1

Average = Total_duration/valid_count

print(f"Valid sessions: {valid_count} | Total duration: {Total_duration} seconds | Average duration: {Average:.0f} seconds")

             

````

## Exercise 5 — Aggregate by category (with missing data)

You have a list of products with their category and units sold.  
Some rows have `None` for `units_sold` — the count was not recorded.

1. Find all unique categories (no imports — just a loop and a list)
2. For each category, calculate the total units sold — use `pass` to skip rows where `units_sold` is `None`
3. Also count how many rows were skipped per category
4. Print a summary, one line per category

**Expected output format**:
```
Electronics     - total units sold: 660    skipped: 1
Furniture       - total units sold:  95    skipped: 0
Lighting        - total units sold:  60    skipped: 1
```

> 💡 Think about how you would do this step by step before writing code. 
> Refer to W14 Reference - a Data Workflow.ipynb if needing info about Generator Expression.

````python
products = [
    {"name": "Laptop",      "category": "Electronics", "units_sold": 120},
    {"name": "Desk Chair",  "category": "Furniture",   "units_sold": 45},
    {"name": "Headphones",  "category": "Electronics", "units_sold": 200},
    {"name": "Bookshelf",   "category": "Furniture",   "units_sold": 50},
    {"name": "Keyboard",    "category": "Electronics", "units_sold": None},
    {"name": "Desk Lamp",   "category": "Lighting",    "units_sold": None},
    {"name": "Monitor",     "category": "Electronics", "units_sold": 140},
    {"name": "Floor Lamp",  "category": "Lighting",    "units_sold": 60},
    {"name": "Webcam",      "category": "Electronics", "units_sold": 200},
]

categories = []

for product in products:
    if product["category"] not in categories:
        categories.append(product["category"])

for category in categories:
    total = 0
    skipped = 0
    for product in products:
        if product["category"] == category:
            if product["units_sold"] is None:
                skipped += 1
            else:
                total += product["units_sold"]
    print(category,total,skipped)
````

## Exercise 6 — String cleaning with regex & ternary expression

You receive a raw export from a sales system. The `revenue` field is a messy string — different formats, currency symbols, dots and commas used inconsistently.

Your job:
1. Use `re.sub()` to strip everything that is not a digit or a dot, then convert to `float`
2. Write the cleaned value back into each dict as a new key `"revenue_clean"`
3. Use a ternary to also add a `"tier"` key: `"Large"` if revenue ≥ 50 000, otherwise `"Small"`
4. Print the cleaned raw_sales

**Expected output format**

```python
{'region': 'North', 'revenue': '85,000 kr', 'revenue_clean': 85000.0, 'tier': 'Large'}
{'region': 'South', 'revenue': 'SEK 42.500', 'revenue_clean': 42500.0, 'tier': 'Small'}
{'region': 'East', 'revenue': '120,000kr', 'revenue_clean': 120000.0, 'tier': 'Large'}
{'region': 'West', 'revenue': '37 800 SEK', 'revenue_clean': 37800.0, 'tier': 'Small'}
{'region': 'Central', 'revenue': 'SEK 61.200', 'revenue_clean': 61200.0, 'tier': 'Large'}
```

> 💡 `re.sub(pattern, replacement, string)` replaces every match of `pattern` with `replacement`.  
> To keep only digits: `re.sub(r'[^\d]', '', value)`  
> In Pandas you will use `Series.str.replace()` and `Series.map()` for the same kind of work — but the logic is identical to what you are writing here.


````python
import re

raw_sales = [
    {"region": "North",   "revenue": "85,000 kr"},
    {"region": "South",   "revenue": "SEK 42.500"},
    {"region": "East",    "revenue": "120,000kr"},
    {"region": "West",    "revenue": "37 800 SEK"},
    {"region": "Central", "revenue": "SEK 61.200"},
]

for sales in raw_sales:
    clean = re.sub(r'[^\d]','',sales["revenue"])
    sales["revenue_clean"] = float(clean)
    sales["tier"] = "Large" if sales["revenue_clean"] > 50000 else "Small"

for sales in raw_sales:
    print(sales)
    
        
````
