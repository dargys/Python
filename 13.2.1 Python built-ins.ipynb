## Data Types

These four data types are also called collection data types or arrays.

### Tuple
An **ordered, immutable** sequence. Stored in round brackets as such `()`.

Use tuples for data that must not change — configuration values, coordinate pairs, fixed column schemas. Functions also return multiple values as tuples.

Read more on [W3Schools](https://www.w3schools.com/python/python_tuples.asp).

````python
# Tuple
dimensions = (1920, 1080)
x, y = dimensions               # unpacking
print(x, y)

# Try below, you will receive TypeError because tuples are immutable
# dimensions[0] = 800           
````

### Lists

A list is an **ordered, mutable** sequence. It can hold any mix of types. Lists are **the most common data container** you will work with before loading data into a DataFrame.

List uses square brackets `[]` and it can be created empty as a placeholder.

Below we'll have a look at indexing & slicing using `[]` and `:` and some common list methods.

#### Indexing and Slicing

Python index starts from 0. You can read about list index() method at [DataCamp](https://www.datacamp.com/tutorial/python-list-index).

````python
scores = [88, 72, 95, 60, 84, 91]

# Indexing and slicing
print(scores[0])       # first element: 88
print(scores[-1])      # last element: 91
print(scores[1:4])     # slice: [72, 95, 60]
print(scores[:3])      # first 3: [88, 72, 95]
print(scores[::2])     # every 2nd: [88, 95, 84]

# show all scores
for s in scores:
    print(s)

# show scores indexes
for s in scores:
    print(scores.index(s))

# or use enumerate() to show indexes next to list
for i, score in enumerate(scores):
    print(i, score) 
````

#### List Methods

> A **method** is a function that belongs to an object. Where a standalone function is called as `function(object)`, a method is called *on* the object using dot notation: `object.method()`. Methods automatically have access to the object they are called on.

When our object example is `scores`, the list method would look like `scores.method()`.

````python
# Modifying
scores.append(78)           # add to end
scores.insert(0, 100)       # insert at position. Here at index 0 (the first position)
scores.remove(60)           # remove first occurrence of value
popped = scores.pop()       # remove and return last element

# Useful list methods
print(len(scores))          # length
print(scores.count(88))     # how many times 88 appears
print(sorted(scores))       # returns a new sorted list
scores.sort(reverse=True)   # sorts in place
print(scores)
````

### Dictionaries

A dictionary stores **key-value pairs** (one pair or many pairs). Since Python 3.7 dictionaries preserve insertion order - that is *append at the end*. 

**Keys must be unique and immutable** (strings and numbers are the most common keys). In a certain sense, we can say *keys* are similar to *column names*.

In Pythonic-flavored data work, a *dict* often represents a single record —  that is one row of named fields. In such a sense, a *list* of *dict* is more comparable to a table that we are more familar with. 

A Dict is stored in curly braces as such `{}`.

Read more on [W3Schools](https://www.w3schools.com/python/python_dictionaries.asp).

````python
# an example of 1 dict

record = {
    "id": 108,
    "product": "Headset",
    "revenue": 5000,
    "in_store": True
}

# Access
print(record["product"])                  # KeyError if missing
print(record.get("margin", 0))            # returns 0 if missing

# Add / update
record["margin"] = 0.35
record["revenue"] = 5500                   # update existing key

# Remove
record.pop("in_store")                     # removes and returns the value

# Iterating — the three common patterns
for key in record:                        # iterate over keys
    print(key)

for value in record.values():             # iterate over values
    print(value)

for key, value in record.items():        # iterate over both
    print(f"{key}: {value}")

# Check if exists
print("revenue" in record)               # True
print("discount" in record)              # False
````

### Set

An **unordered collection of <font color=MediumOrchid>unique</font> elements**. Sets are optimised for membership testing and removing duplicates. This make it very handy in data analysis.

> So what type of brackets store a set?

That will be `{}`. Good to know that set() always returns a set object — it never preserves the original type, meaning if we apply set() on a list, the returned results will be a set, not a list. 

You can read more about **set** at [W3schools](https://www.w3schools.com/python/python_sets.asp).

````python
# Set
departments_raw = ["Sales", "HR", "Sales", "Finance", "HR", "HR"]
departments = set(departments_raw)          # duplicates removed
print(departments)                          # order is not guaranteed

# Fast membership test — O(1) regardless of set size
print("Sales" in departments)   # True

# Set operations — useful for comparing datasets
team_a = {"Alice", "Bob", "Carol"}
team_b = {"Bob", "Carol", "Dave"}

print(team_a & team_b)          # intersection: {'Bob', 'Carol'}
print(team_a | team_b)          # union: all members
print(team_a - team_b)          # difference: only in team_a
````

## Comprehensions

Comprehensions are a concise, readable way to build new lists, dicts, or sets from existing data — with optional filtering. They replace many simple `for` loops in idiomatic Python. This has a more Pythonic flavor.

> Remember the language analogy in lecture 1? <br>
> In a similar way we can say:<br>
> **comprehension** is a noun - it creates a new thing (a list, dict...)<br>
> **`for` loop** is like a verb to do something repetitively

**General form:**

```python
[expression for item in iterable if condition]
```
The `if condition` part is optional.

Comprehensions are very useful so it worths spending time reading and practicing them.

Below are some common patterns we can use.

````python
# We have a list with a number of dicts as element

products = [
    {"name": "Widget A", "revenue": 50000, "category": "Hardware"},
    {"name": "Widget B", "revenue": 12000, "category": "Software"},
    {"name": "Widget C", "revenue": 95000, "category": "Hardware"},
    {"name": "Widget D", "revenue": 3000,  "category": "Software"},
]
````

#### List Comprehension

````python
# All product names
names = [p["name"] for p in products]
print(names)

# Only high-revenue products
high_value = [p["name"] for p in products if p["revenue"] > 20_000]
print(high_value)

# Transform: apply a 10% uplift to all revenues
uplifted = [p["revenue"] * 1.1 for p in products]
print(uplifted)
````

#### Dict comprehension

````python
# Build a name → revenue lookup
revenue_map = {p["name"]: p["revenue"] for p in products}
print(revenue_map)

# Only Hardware products in the lookup
hardware_map = {p["name"]: p["revenue"] for p in products if p["category"] == "Hardware"}
print(hardware_map)
````

### Set Comprehension

````python
# Distinct categories
categories = {p["category"] for p in products}
print(categories)
````

Real-world data is full of messy strings. Python's string methods handle the most common cleaning tasks without any library.


Below are some examples applying string methods on a variable. Modify the variable and see if results are as you expecte them.

### Cleaning & format

````python
raw = "  Widget A — Revenue: 50,000 EUR  "

print(raw.strip())              # remove leading/trailing whitespace
print(raw.lower())
print(raw.upper())
````

### Search & replace

````python
# Search
print("Revenue" in raw)         # True
print(raw.startswith("  "))     # True
print(raw.find("Revenue"))      # index position of match

# Replacing
cleaned = raw.strip().replace(",", "").replace("EUR", "").strip()
print(cleaned)
````

### Split, join & parse

````python
# Splitting — turn a string into a list
csv_line = "North,88,Hardware,Active"
fields = csv_line.split(",")
print(fields)                   # ['North', '88', 'Hardware', 'Active']

# Joining — turn a list back into a string
print(" | ".join(fields))

# Parsing a number from text
amount_str = "50000"
amount = float(amount_str)
print(amount * 1.1)
````

## Some Built-in Functions for Data

These are always available — no import needed. Together they cover most of what you need for quick data summarisation and transformation.

Read about more [Python built-in functions at Python.org](https://docs.python.org/3/library/functions.html)

Below are some examples for you to try out.

### Numericals

````python
values = [3, 8, 1, 6, 4, 9, 2, 7, 5]

print(len(values))             # 9
print(sum(values))             # 45
print(min(values))             # 1
print(max(values))             # 9
print(round(sum(values) / len(values), 2))  # average: 5.0
````

### Sorting

````python
print(sorted(values))                      # new sorted list, ascending
print(sorted(values, reverse=True))        # descending
````

````python
# Sort a list of dicts by a field

records = [{"name": "C", "score": 85}, {"name": "A", "score": 92}, {"name": "B", "score": 78}]

by_score = sorted(records, key=lambda r: r["score"], reverse=True)
print(by_score)
````

### `format()`

Sometimes it's necessary to beautify for example number format. This function returns string type. Below are some examples with commonly used number format.

If we run `help(format)`, we will see:

```
format(value, format_spec='', /)
    Return type(value).__format__(value, format_spec)
```

If very interested in reading about `format_spec` mentioned above, one can read [Python doc: format specification mini language](https://docs.python.org/3/library/string.html#format-specification-mini-language).

For a hands-on practical guide, please refer to [DataCamp - Python f-string: A Complete Gudie](https://www.datacamp.com/tutorial/python-f-string).

````python
sales = 50000
units = 10

print(f"Sales: {sales:,}")                      # Sales: 50,000
print(f"Sales: {sales:,.2f}")                   # Sales: 50,000.00
print(f"Sales: {int(sales/units):>10,}")        # Sales:      5,000 shows right aligned result
print(f"Growth: {0.1374:.1%}")                  # Growth: 13.7%
````

````python
# Try below example and see results

reps = [("Alice", 15000), ("Bob", 8200), ("Carol", 23000)]

for name, sales in reps:
    print(f"{name:<10} SEK {sales:>10,.0f}")
````

### `enumerate()`

This function allows you to loop through an iterable (list, tuple, string) and have access to both the index and the element itself.

For further use of `enumerate()`, read on [Real Python](https://realpython.com/python-enumerate/).

````python
# again, we can use enumerate() — index + value

regions  = ["North", "South", "East"]

for i, region in enumerate(regions, start=1):
    print(f"{i}. {region}")
````

### `any() / all()`

These pass logical tests across a collection. Examples below:

````python

scores = [88, 72, 95, 60, 84]
print(all(s >= 50 for s in scores))   # True — everyone passed
print(any(s >= 90 for s in scores))   # True — at least one distinction
````
