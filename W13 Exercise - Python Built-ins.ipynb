### Exercise 1.2 — Sets

### Exercise 2.1 — Calculate revenue

Using a **list comprehension**, create a list called `revenues` where each element is `units * unit_price` for the corresponding record in `sales_records`.

Then use built-in functions to print:
- Total revenue across all reps
- Minimum revenue
- Maximum revenue
- Average revenue (rounded to 2 decimal places)

**Expected output (example format):**
```
Total revenue:   SEK 399,100
Min revenue:     SEK 18,000
Max revenue:     SEK 144,500
Avg revenue:     SEK 49,887.50
```

````python
# YOUR CODE HERE
revenues = [r["units"] * r["unit_price"] for r in sales_records]

total = sum(revenues)
avg = total / len(revenues)

print(f"Total revenue: {total:,}")
print(f"Min revenue: {min(revenues):,}")
print(f"Max revenue: {max(revenues):,}")
print(f"Avg revenue: {avg:,.2f}")
````

### Exercise 2.2 — Filter with comprehension

Using a **list comprehension with a condition**, create a list called `high_performers` that contains only the **rep names** whose revenue (units × unit_price) is **above 50,000**.

Print the list and the count.

**Expected output (example format):**
```
High performers (3): ['Alice', 'Grace', 'Carol']
```
*(Names and order may vary depending on your implementation.)*

````python
# YOUR CODE HERE
high_performers = [r["rep"] for r in sales_records if r["units"] * r["unit_price"] > 50000]

print(f"High performers({len(high_performers)}): {high_performers}")
````

### Exercise 2.3 — Dict comprehension

Using a **dict comprehension**, build a lookup dictionary called `rep_revenue` where:

- Value = their total revenue (int)

Print the dictionary.
