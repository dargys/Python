### pip/shell vs. help()

**The difference below the two approaches are**:

- `%pip show pandas` is a **pip/shell command** that shows package metadata — version, author, install location, dependencies, etc. 
- `help('pandas')` uses **Python's built-in help system** and shows the module's docstring and public API — classes, functions, and their documentation.

More specifically:

- `%pip show pandas` reads from the package's dist-info **metadata** folder (e.g. `.venv/lib/pandas-2.x.x.dist-info/`). This is pure packaging metadata written by pip at install time — it knows nothing about the code itself.
- `help('pandas')` reads from the actual **source code** inside `.venv/lib/pandas/` — specifically the docstrings embedded in the .py files.

Now open your `.venv` folder and look for the following.

```bash
.venv/lib/
├── pandas/                  ← help() reads from here (source + docstrings)
│   ├── __init__.py
│   ├── core/
│   └── ...
└── pandas-2.x.x.dist-info/ ← pip show reads from here (install metadata)
    ├── METADATA
    ├── RECORD
    └── ...
```

> Now you know **"where does pandas live?" in this `.venv`** on your laptop.

## What's Inside a Package

Below we will use `pandas` as an example.

### 1. `dir()` lists all inside

We start with `dir()` which lists all attributes and methods inside a package.

````python
import pandas as pd

dir(pd)
````

````python
# a cleaner look using for-loop

for x in dir(pd):
    print(x)
````

````python
# an oneliner list comprehension if one condition

pandas_read = [x for x in dir(pd) if x.startswith("read")]
pandas_read
````

### 2. `inspect` module 

`dir()` returns everything (classes, constants, submodules etc.). If you want to see only functions or methods, you can use built-in `inspect` module.

````python
# First let's have a look at inspect module

help("inspect")
````

````python
import inspect
import pandas as pd

# Use inspect.getmembers(obj) to view pandas top 5 elements
inspect.getmembers(pd)[0:5]
````

````python
# Above output shows the returned output is a list with tuples. Each tuple contains 2 elements, which are name, obj in below list comprehension

functions = [name for name, obj in inspect.getmembers(pd) if inspect.isfunction(obj)]
functions
````
