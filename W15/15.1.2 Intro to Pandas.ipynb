### NumPy ndarrays

**ndarray** stands for N-dimensional array object in NumPy. This feature enables larger scale mathematical calculation.  

```python
import numpy as np

# create and show 1D array
data1 = [6, 7.5, 8, 0, 1]
arr1 = np.array(data1)

arr1
```

```python
# create and show 2D array, note that it's without labels
data2 = [[1, 2, 3, 4], [5, 6, 7, 8]]
arr2 = np.array(data2)

# take a look at it
arr2
```

```python
print(type(arr2))  # check type
print(arr2.ndim)  # how many dimensions?
print(arr2.shape)  # array in what shape (for 2D, it is as in row x column)?
```

```python
print(arr2)
print(f"There are {arr2.ndim} dimensions in the array above")
print(f"It has data type: {arr2.dtype}")
print(
    f"This array has a shape of {arr2.shape[0]} x {arr2.shape[-1]} as it has {arr2.shape[0]} rows x {arr2.shape[-1]} columns"
)
```

```python
# Methematical calculation on the array

arr2 * 3
```

### ndarray Data Type 

One array can be created with an explicitly specified data type `dtype`. However, one cannot mix data types inside an array. When ndarray is created, it can only hold a single data type.

```python
# compare the array datatypes

print(arr1.dtype)
print(arr2.dtype)
```

```python
arr_mix = np.array([[1, 2, 3, 4], ["a", "b", "c", "d"]])

arr_mix.dtype  # will return dtype('<U21') which indicates it's string of a certain length.
```

```python
# When we call arr_mix, we can see the integers are coerced into string type.

arr_mix
```

```python
# You can cast ndarray datatype with astype. Below we temporarily cast arr2 into a float64 type. See the decimal . indicating data type float?

print(arr2.astype(np.float64))
```

```python
# Compare to arr2 as is.

arr2
```

### 1. The two core objects: Series and DataFrame

#### Series: 1D + label

A **Series** is a one-dimensional labelled array — think of it as a single column with an index and you can customize the index. Well, yes, this might be a bit confusing at the start.

Let's start from the more familar standard index as 0, 1, 2, ...

How we create a pandas.Series might remind you something similar: a Python list. The difference is that a Series naturally carries an **index** (labels for each position) without you calling it explicitly and it knows its **dtype** (the data type of its values - just like what we've seen in NumPy basics above).

Run code block below and observe returned output.

```python
import pandas as pd

# A Series from a list
revenues = pd.Series([182000, 95000, 28000, 61000])
revenues
```

```python
# Below are some of the pandas.Series.method you can use to call specific info. You can run help(pandas.Series.method) in Python REPL to read about these methods.

# print is convenient to see all output in one-go. It's also interesting to observe the output if no print is used.

print("dtype:", revenues.dtype)
print("values:", revenues.values)  # the underlying array
print("index:", revenues.index)  # the labels (0, 1, 2, 3 by default)
```

```python
revenues.dtype
```

```python
revenues.values
```

```python
revenues.index
```

While you can use the default index 0, 1, ..., it is also possible to label element in the Series as below

```python
# A Series with a custom index
revenues = pd.Series(
    [182000, 95000, 40040, 59780],
    index=["Pasta Palace", "Burger Barn", "Sushi Spot", "Taco Town"],
)
revenues
```

```python
# Above may look like a table, but not yet.

type(revenues)

# It's only rows got row label (customized index). This is still only 1D data without a column label.
```

```python
# try again to see the index

revenues.index # now with the labels
```

### 4. Selecting columns and rows

#### Column selection

A single column returns a **Series**. Multiple columns return a **DataFrame**.

This will feel familiar — selecting a key from a dict works the same way.

```python
# Single column — returns a Series

df["restaurant"]
```

```python
type(df["restaurant"])
```

```python
# Multiple columns — returns a DataFrame and often a subset

# Note the double brackets: 
# outer [] selects from df
# inner [] is the list of column names

df[["restaurant", "city", "revenue_sek"]]
```

#### Row selection with `.loc` and `.iloc`

| Method | Selects by | Example |
|--------|-----------|----------|
| `.loc` | **label** (index value or column name) | `df.loc[0]`, `df.loc[0, "city"]` |
| `.iloc` | **integer position** (0-based) | `df.iloc[0]`, `df.iloc[0, 1]` |

When the index is 0, 1, 2... (the default), `.loc` and `.iloc` look the same — but they're not. 

> **Rule of thumb**: use `.iloc` when you mean "position", `.loc` when you mean "label".

```python
# let's have a look at the DataFrame again, pay attention to the index per row

df
```

```python
df.iloc[0]  # select a row by position: index 0 (the first row)
```

```python
df.loc[0]   # identical result is returned because we haven't labeled our index
```

```python
# One can also choose to see a row as a row - this work for tables that are not so wide.

df.iloc[0:1]    
```

```python
# Slicing — first three rows

df.iloc[0:3]
```

```python
# Select a specific cell: row 2, column "city"

df.loc[2, "city"]
```

```python
# same cell by position

df.iloc[2, 1]
```

### 5. Cleaning data — the pandas way

In earlier lectures, we did quite a few *manual* cleaning by writing functions and loops from Python built-ins. Pandas has built-in tools for most of these jobs.

Let's reload the *raw* data and clean it with pandas this time.

```python
# Raw data — same as Lecture 4
raw_data = [
    {"restaurant": "  pasta palace   ",  "city": "Stockholm",  "revenue": "182000",    "currency": "SEK", "covers": "210", "rating": "4.5"},
    {"restaurant": "Burger Barn  ",   "city": "Stockholm",  "revenue": "95000",     "currency": "SEK", "covers": "180", "rating": "3.8"},
    {"restaurant": "SUSHI SPOT",      "city": "Copenhagen", "revenue": "28000",     "currency": "DKK", "covers": "95",  "rating": "4.9"},
    {"restaurant": "taco town",       "city": " Oslo ",       "revenue": "61000",     "currency": "NOK", "covers": "140", "rating": None},
    {"restaurant": "Pasta Palace",    "city": "Stockholm",  "revenue": "182000",    "currency": "SEK", "covers": "210", "rating": "4.5"},
    {"restaurant": "Green Garden",    "city": "Oslo",       "revenue": "see notes", "currency": "NOK", "covers": "88",  "rating": "4.1  "},
    {"restaurant": "Burger Barn",     "city": "Copenhagen", "revenue": "31500",     "currency": "DKK", "covers": "210", "rating": "6.0"},
    {"restaurant": "  Ramen House ",  "city": "Stockholm",  "revenue": "74000",     "currency": "SEK", "covers": None,  "rating": "4.2"},
]

raw = pd.DataFrame(raw_data)
raw
```

#### 5a. Fix restaurant names

In pandas: `Series.str` — a string accessor that applies string methods to every value at once.

```python
# .str.strip() removes leading/trailing whitespace from every value
# .str.title() applies title case to every value
# Chain them together — just like chaining string methods in Python

raw["restaurant"] = raw["restaurant"].str.strip().str.title()
raw["restaurant"]

# Compare to function written with built-ins:
# def clean_name(name):
#     return name.strip().title()
# [clean_name(r["restaurant"]) for r in raw_data]
```

#### About method chaining

The above method chaining is very useful and concise. However, when it becomes too long, it will be hard to read. 

Therefore, below code formatting makes the chaining easier to read. But it is important to test your code so the breaks are correct wrapped without breaking code.

```python
# An alternative way to chain methods

raw["restaurant"] = (raw["restaurant"]
                    #comment
                    .str.strip()
                    #comment
                    .str.capitalize())

raw["restaurant"]
```

#### 5b. Convert revenue to numeric

In earlier lecture, we used: `parse_number()` — `try/except` around `float(value)`  
In pandas: `pd.to_numeric()` with `errors='coerce'` — bad values become `NaN` automatically.

```python
# check data types before cleansing

raw[["revenue", "covers", "rating"]].dtypes
```

```python
# errors='coerce' turns anything that can't be converted into NaN
# 'see notes' becomes NaN — no try/except needed

raw["revenue"] = pd.to_numeric(raw["revenue"], errors="coerce")
raw["covers"]  = pd.to_numeric(raw["covers"],  errors="coerce")
raw["rating"]  = pd.to_numeric(raw["rating"],  errors="coerce")
```

```python
# After cleansing

raw[["revenue", "covers", "rating"]].dtypes
```

```python
raw[["revenue", "covers", "rating"]]
```

#### 5c. Handle missing values

In earlier lecture: `if value is None: return default`  
In pandas: `.fillna()` to replace NaN with a default, or `.dropna()` to remove rows with missing values.

```python
# Replace NaN revenue with 0 (same default we used in parse_number)
raw["revenue"] = raw["revenue"].fillna(0)

raw[["revenue", "rating"]]
```

```python
# For rating: cap values above 5.0 — then leave NaN as-is (no valid rating to fall back on)
raw["rating"] = raw["rating"].clip(upper=5.0)

# Observe row index 6
raw[["revenue", "rating"]]
```

#### 5d. Remove duplicates

In pandas: `.drop_duplicates()` — one line.

```python
# After cleansing, below rows appear to be identical

raw.iloc[[0,4]]
```

```python
# Keep first occurrence

raw = raw.drop_duplicates(subset=["restaurant", "city", "revenue", "currency", "covers","rating"], keep="first")
```

```python
# After dedup

raw
```
