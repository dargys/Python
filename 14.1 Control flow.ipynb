## 1. `if / elif / else`

Python checks conditions top to bottom and runs the **first** block that is `True`. 

```python
if condition:
    # runs if condition is True
elif other_condition:
    # runs if the first was False, this is True
else:
    # runs if none of the above were True
```


````python
temperature = 18

if temperature >= 25:
    print("Warm — maybe shorts weather.")
elif temperature >= 10:
    print("Mild — a light jacket should do.")
else:
    print("Cold — dress up properly.")
````

````python
# You can chain as many elif branches as you need
day = "Monday"

if day == "Monday":
    print("Start of the week.")
elif day == "Wednesday":
    print("Halfway there.")
elif day == "Friday":
    print("Weekend soon!")
else:
    print(f"{day} — just another day.")
````

### Combining conditions with `and`, `or`, `not`

````python
is_raining = False
has_umbrella = False

if is_raining and not has_umbrella:
    print("You're going to get wet.")
elif is_raining and has_umbrella:
    print("You're prepared.")
else:
    print("No rain, no worries.")
````

## 2. `for` loops

A `for` loop iterates over any *sequence* — a list, a string, a range, etc.

```python
for item in sequence:
    # runs once for each item
```

````python
cities = ["Stockholm", "Gothenburg", "Malmö", "Uppsala"]

for city in cities:
    print(f"Hello from {city}!")
````

````python
# range() when you need a counter
for n in range(1, 6):
    print(f"Step {n}")
````

````python
# Looping over a list of dicts — each dict is like a row of data
people = [
    {"name": "Noah",  "age": 28},
    {"name": "Liam", "age": 17},
    {"name": "Alma", "age": 34},
]

for person in people:
    status = "adult" if person["age"] >= 18 else "minor"
    print(f"{person['name']} is {person['age']} — {status}")
````

````python
# enumerate() gives you index + value at the same time
fruits = ["apple", "banana", "cherry"]

for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")
````

````python
# meet condition then add to new list
fruits = ["apple", "banana", "cherry"]
fruit_delivered = "apple"
fruit_to_eat = []

for fruit in fruits:
    if fruit_delivered == fruit:
        fruit_to_eat.append(fruit)
````

## 3. `while` loops

A `while` loop keeps going as long as a condition is `True`.  
Use it when you don't know in advance how many iterations you need.

```python
while condition:
    # keeps running as long as condition is True
```

````python
# Count down from 5
countdown = 3

while countdown > 0:
    print(countdown)
    countdown -= 1

print("Go!")
````

````python
# Keep trying until we get a valid answer
# (simulating responses rather than using input())
responses = ["maybe", "", "yes"]
accepted = ["yes", "no"]

i = 0
answer = responses[i]

while answer not in accepted:
    print(f"'{answer}' is not valid. Try again.")
    i += 1
    answer = responses[i]

print(f"Got a valid answer: '{answer}'")
````

> ⚠️ <font color=DeepPink>**Watch out for infinite loops!** Always make sure the condition can eventually become `False`, or use a `break`.</font>

## 5. `break` and `continue`

- **`break`** — exit the loop immediately
- **`continue`** — skip the rest of this iteration and go to the next one

```python
for item in sequence:
    if condition:
        break      # exit the loop entirely
    if other_condition:
        continue   # skip to the next iteration
```

````python
# break: stop as soon as we find what we're looking for
names = ["Alice", "Bob", "Charlie", "Diana", "Eve"]

for name in names:
    if name == "Charlie":
        print(f"Found Charlie! Stopping.")
        break
    print(f"Not Charlie: {name}")
````

````python
# continue: skip missing values, process the rest
values = [10, None, 5, None, 8, 3]

total = 0
skipped = 0

for v in values:
    if v is None:
        skipped += 1
        continue
    total += v

print(f"Total: {total}  |  Skipped: {skipped} missing values")
````

## 6. Ternary expressions (inline `if`)

A compact way to write a simple if/else on a single line:

```python
value_if_true  if  condition  else  value_if_false
```

or even simpler:

```python
value = x  if condition  else  y
```

````python
age = 20
label = "adult" if age >= 18 else "minor"
print(label)
````

````python
# Works well inside a loop
scores = [88, 45, 72, 33, 91, 60]

for score in scores:
    result = "Pass" if score >= 70 else "Fail"
    print(f"{score}  →  {result}")
````

````python
# And inside a list comprehension — a very common Python pattern
results = ["Pass" if score >= 50 else "Fail" for score in scores]
print(results)
````

💡 That last cell combines a **ternary expression** with a **list comprehension**. You'll see this pattern a lot once we get into Pandas — for now, just notice that it exists.

## A First Look at Data Workflow

With these basics (before pandas and other packages), you can already do meaningful data work using just lists of dicts, comprehensions, and built-ins. This is also the mental model pandas builds on.

Here is a small end-to-end example: load, clean, filter, aggregate, and report.

💡 Before we formally work through Functions ``def`, can you see what problem below data workflow might have?

````python
# --- Raw data (as you might receive from an API or CSV) ---
raw_data = [
    {"region": "North", "product": "Widget A", "revenue": "120000", "cost": "80000"},
    {"region": "South", "product": "Widget B", "revenue": "85000",  "cost": "60000"},
    {"region": "North", "product": "Widget C", "revenue": None,     "cost": "40000"},
    {"region": "East",  "product": "Widget A", "revenue": "200000", "cost": "130000"},
    {"region": "South", "product": "Widget C", "revenue": "30000",  "cost": "20000"},
]

# --- Step 1: Parse & clean ---
def parse_record(r):
    try:
        rev = float(r["revenue"])     # convert string to number
        cost = float(r["cost"])
        return {
            "region":  r["region"],
            "product": r["product"],
            "revenue": rev,
            "cost":    cost,
            "profit":  rev - cost,
            "margin":  (rev - cost) / rev
        }
    except (TypeError, ValueError):
        return None   # skip rows with missing/bad data

clean = [parse_record(r) for r in raw_data]
clean = [r for r in clean if r is not None]   # drop failed rows

print(f"Rows after cleaning: {len(clean)}")

# --- Step 2: Filter ---
high_margin = [r for r in clean if r["margin"] >= 0.30]
print(f"High-margin rows: {len(high_margin)}")

# --- Step 3: Aggregate by region ---
regions = {r["region"] for r in clean}

for region in sorted(regions):
    rows = [r for r in clean if r["region"] == region]
    total_rev = sum(r["revenue"] for r in rows)
    total_profit = sum(r["profit"] for r in rows)
    print(f"{region:8s}  Revenue: {total_rev:>10,.0f}  Profit: {total_profit:>10,.0f}")

# --- Step 4: Top product by revenue ---
top = sorted(clean, key=lambda r: r["revenue"], reverse=True)[0]
print(f"\nTop product: {top['product']} in {top['region']} — {top['revenue']:,.0f}")
````

### Mutable vs Immutable Objects

This is one of the most common sources of unexpected bugs for newcomers. Understanding it now saves a lot of debugging later.

**Mutable** objects (`list`, `dict`, `set`) can be changed after creation.  
**Immutable** objects (`int`, `float`, `str`, `tuple`) cannot.

The subtle problem: assigning a mutable object to a new variable does **not** copy it. Both variables point to the **same object** in memory.

````python
# --- The problem ---
original = [10, 20, 30]
alias = original        # NOT a copy — both point to the same list

alias.append(99)
print(original)         # [10, 20, 30, 99] — original was changed too!

# --- The fix: copy explicitly ---
original = [10, 20, 30]
copy = original.copy()  # or: list(original)  or  original[:]

copy.append(99)
print(original)         # [10, 20, 30]  — untouched
print(copy)             # [10, 20, 30, 99]

# Same issue with dicts
rec = {"name": "Alice", "score": 88}
rec_copy = rec.copy()
rec_copy["score"] = 0
print(rec)              # original unchanged

# Immutable: no such issue
a = "hello"
b = a
b = b.upper()           # creates a NEW string, does not modify a
print(a)                # 'hello' — unchanged
````

### `*args` and `**kwargs`

You will encounter these in library documentation and in functions designed to accept a flexible number of arguments.

- `*args` — collects any number of **positional** arguments into a **tuple**
- `**kwargs` — collects any number of **keyword** arguments into a **dict**

The names `args` and `kwargs` are conventions — the `*` and `**` are what matter.

````python
# *args — variable number of positional arguments
def total(*amounts):
    return sum(amounts)

print(total(100, 200))           # 300
print(total(10, 20, 30, 40))     # 100

# **kwargs — variable number of keyword arguments
def describe_record(**fields):
    for key, value in fields.items():
        print(f"  {key}: {value}")

describe_record(name="Widget A", revenue=50_000, region="North")

# In practice: you'll see these when calling pandas, matplotlib, etc.
# e.g. df.plot(kind="bar", title="Revenue", figsize=(10,5), color="steelblue")
# All those extra keyword arguments go into **kwargs inside the function.
````
