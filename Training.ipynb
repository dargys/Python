```python
def apply_discount(price, rate):
    saving = price * rate
    final = price - saving
    return final

cost = apply_discount(200, 0.15)
print(cost)
```

```python
employees = [
    {"name": "Alice", "salary": 80000, "department": "Engineering"},
    {"name": "Bob",   "salary": 75000, "department": "Marketing"},
    {"name": "Carol", "salary": 90000, "department": "Engineering"},
]

results = []

for employee in employees:
    if employee["department"] == "Engineering":
        employee["new_salary"] = employee["salary"]*1.1
        results.append(employee)

print(results)
```

```python
products = [
    {"name": "Laptop", "price": 999, "in_stock": True},
    {"name": "Mouse",  "price": 25,  "in_stock": False},
    {"name": "Keyboard", "price": 79, "in_stock": True},
]

available = []
for p in products:
    if p["in_stock"]:
        if p["price"] < 100:
            available.append(p["name"])

print(available)
```

```python
def summarise(data):
    total = sum(data)
    average = total / len(data)
    return {"total": total, "average": average}

monthly_sales = [12000, 15000, 9000, 18000, 14000]
report = summarise(monthly_sales)
print(report["average"])
```

```python
orders = [
    {"item": "Book",   "qty": 3, "price": 12.99},
    {"item": "Pen",    "qty": 10, "price": 1.49},
    {"item": "Notebook", "qty": 2, "price": 5.99},
]

total_revenue = 0
for o in orders:
    line_total = o["qty"] * o["price"]
    if line_total > 20:
        total_revenue = total_revenue + line_total

print(total_revenue)
```
