```python
def summarise(data):
    total = sum(data)
    average = total / len(data)
    return {"total": total, "average": average}

monthly_sales = [12000, 15000, 9000, 18000, 14000]
report = summarise(monthly_sales)
print(report["average"])
```

```python
# Keep trying until we get a valid answer
# (simulating responses rather than using input())
responses = ["maybe", "", "yes"]
accepted = ["yes", "no"]

i = 0
answer = input("Are you comming?")

while answer not in accepted:
    print(f"'{answer}' is not valid. Try again.")
    i += 1
    answer = responses[i]

print(f"Got a valid answer: '{answer}'")
```

```python
# --- Raw data (as you might receive from an API or CSV) ---
raw_data = [
    {"region": "North", "product": "Widget A", "revenue": "120000", "cost": "80000"},
    {"region": "South", "product": "Widget B", "revenue": "85000",  "cost": "60000"},
    {"region": "North", "product": "Widget C", "revenue": None,     "cost": "40000"},
    {"region": "East",  "product": "Widget A", "revenue": "200000", "cost": "130000"},
    {"region": "South", "product": "Widget C", "revenue": "30000",  "cost": "20000"},
]
```

```python
# --- Step 1: Parse & clean ---
def parse_record(r):
    try:
        rev = float(r["revenue"])     # convert string to number
        cost = float(r["cost"])
        return {
            "region":  r["region"],
            "product": r["product"],
            "revenue": rev,
            "cost":    cost,
            "profit":  rev - cost,
            "margin":  (rev - cost) / rev
        }
    except (TypeError, ValueError):
        return None   # skip rows with missing/bad data

```

```python
clean = [parse_record(r) for r in raw_data]
```

```python
print(len(clean))
```

```python

clean = [r for r in clean if r is not None]   # drop failed rows


print(f"Rows after cleaning: {len(clean)}")
```

```python
from pathlib import Path

def print_tree(path, prefix=""):
    '''
    print a given folder's entire structure and underlyting files. 
    '''
    contents = sorted(Path(path).iterdir())
    for i, item in enumerate(contents):
        connector = "└── " if i == len(contents) - 1 else "├── "
        print(prefix + connector + item.name)
        if item.is_dir():
            extension = "    " if i == len(contents) - 1 else "│   "
            print_tree(item, prefix + extension)
```
