### 1. Language Characteristics

Python is an **interpreted, high-level, general-purpose** programming language. Its core features include:

- **Dynamic Typing**: Variable types are determined at runtime based on the assigned value — no need to declare a type in advance. This improves development flexibility.
- **Interpreted Language**: Code is interpreted and executed line-by-line at runtime, skipping a compilation step and speeding up the development-test cycle.

### 4. Strings & f-strings

Strings can use single or double quotes. 

For formatting values into strings, use **f-strings** (Python 3.6+): prefix the string with `f` and place any expression inside `{}`. It is an easy way to generate strings that contain interpolated expressions. Any code between curly braces {} will be evaluated and then the result will be turned into a string and intserted into the overall string. 

F-strings are really useful. Try with below examples. Read more at [W3schools.com](https://www.w3schools.com/python/python_string_formatting.asp) 

````python
product = "Headset"
revenue = 7576.725
growth = 0.0523

# Basic embedding
print(f"Product: {product}")            # Product: Headset

# Number formatting
print(f"Revenue: {revenue:,.2f}")       # Revenue: 7,576.73
print(f"Growth:  {growth:.1%}")         # Growth:  5.2%

# Any expression works inside {}
print(f"Double revenue: {revenue * 2:,.0f}")    # Double revenue: 15,153
````

### 6. Conditional Statements & Control Flow

Python by default evaluates every line of code from top to bottom, sequentially. The order in which Python runs our code is called **control flow**, and Python provides functionality we can use to change the control flow. 

The first tool Python provides for us is the **if-statement**. 

We write an [**if-statement**](https://www.w3schools.com/python/gloss_python_if_statement.asp) like this:

```python
    if bool:
        code_1
    else:
        code_2
```

`bool` here is a boolean expression, evaluating to either **`True`** or **`False`**

Think of this as a fork in the road. Python will read the boolean at the top, then if it is **`True`**, it will execute **`code_1`** not **`code_2`**

If it is **`False`**, it will skip **`code_1`** and only execute **`code_2`**

Read more about control flow on [Real Python](https://realpython.com/python-control-flow/).

````python
age = 19

if age >= 20:
    print("Come on in")
else:
    print("Go home kid")
````

For multiple conditions, one can use `if-elif-else`.

```python
    if condition_1:
        code_1
    elif condition_2:
        code_2
    else:
        code_3
```

````python
score = 85

if score >= 90:         
    grade = 'A'
elif score >= 80:
    grade = 'B'
else:
    grade = 'C'

print(f"Score: {score} → Grade: {grade}") # 
````

### 7. Loops

- **`for` loop**: iterates over an iterable (list, string, `range()`, etc.).

```python
for item in iterable:
    statement
```

- **`while` loop**: repeats while a condition is `True`. 

```python
while expression:
    statement
```

Read more on [W3Schools](https://www.w3schools.com/python/python_for_loops.asp).

````python
# for loop over a range
for i in range(4):       # 0, 1, 2, 3
    print(i, end=" ")
print()

# for loop over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

# enumerate() — index + value together
for i, fruit in enumerate(fruits, start=1):
    print(f"{i}. {fruit}")

# while loop
count = 5
while count > 0:        # as long as count is larger than 0
    count -= 1          # reduce 1
    print(count)
````

### 8. Functions

Functions are defined with `def`. They can have default argument values, return multiple values (as a tuple), and include a docstring explaining their purpose.

**Scope:** variables created inside a function are **local** — they only exist within that function and are gone when it returns.

Read more on [W3Schools](https://www.w3schools.com/python/python_functions.asp).

````python
def greet(name, greeting="Hello"):
    """Return a greeting string. 'greeting' has a default value."""
    return f"{greeting}, {name}!"

print(greet("Alice"))                # uses default
print(greet("Bob", greeting="Hi"))   # override default

# Returning multiple values
def min_max(values):
    return min(values), max(values)

lo, hi = min_max([3, 1, 4, 1, 5, 9])
print(f"Min: {lo}, Max: {hi}")

# Lambda — a compact anonymous function
double = lambda x: x * 2
print(double(7))   # 14
````

### 9. Methods

A **method** is a function that belongs to an object. Where a standalone function is called as `function(object)`, a method is called *on* the object using dot notation: `object.method()`. Methods automatically have access to the object they are called on.

Every data type in Python comes with its own set of built-in methods, such as:

- str.upper()
- str.split()
- list.append()
- list.sort()
- dict.keys()

Below are a few examples. 

The key distinction:
- `len(my_list)` — built-in **function**, takes the object as an argument
- `my_list.append(4)` — **method**, called on the object itself

You will encounter this pattern constantly in Lecture 2 (`str.upper()`,`str.split()`,`list.append()`, `list.sort()`) and later throughout pandas (`df.head()`, `df.groupby()`, `series.mean()`, etc.).

Read more on [W3Schools](https://www.w3schools.com/python/python_lists_methods.asp).

````python
# String methods
text = "  hello, world  "
print(text.strip())           # remove whitespace: 'hello, world'
print(text.upper())           # 'HELLO, WORLD'
print(text.replace(",", ";")) # 'hello; world'

# List methods
numbers = [3, 1, 4, 1, 5]
numbers.append(9)             # modifies the list in place
numbers.sort()                # modifies the list in place
print(numbers)
````

### 11. Indentation & Code Blocks

Python uses indentation to show **code hierarchy** — and it's not just formatting, it's logic and a clear visual structure. All statements in the same code block must use the same indentation level — the standard is **4 spaces**.

Here is an example:

```python
if score > 50:
    print("Pass")       # only runs if score > 50
    if score > 90:
        print("Distinction")  # only runs if score > 90
print("Done")           # always runs
```

- Each indentation level = one more layer of nesting
- Reading tip: follow the indentation to understand *what controls what*

````python
# Correct indentation
if True:
    print("True")
else:
    print("False")

# ❌ Inconsistent indentation causes an IndentationError:
# if True:
#     print("Answer")
#     print("True")
# else:
#     print("Answer")
#   print("False")    # Wrong indentation — runtime error
````

####  `None` & `NoneType`

`None` represents the **absence of a value**. It is its own type (`NoneType`) and is Python's equivalent of a null. A few important rules:

- Always check for `None` using `is None`, not `== None`
- `None` is not `0`, not `False`, not an empty string — it is uniquely itself
- A function that reaches its end without a `return` statement implicitly returns `None`

Read more on [W3Schools](https://www.w3schools.com/python/python_none.asp).

````python
result = None

print(result is None)     # True  — correct way to check
print(result == False)    # False — None is not False
print(result == 0)        # False — None is not zero
print(type(result))       # <class 'NoneType'>
````

### Execution Order and State

Above we looked at programming concepts with a language analogy, I hope that's helpful. 

However, in code, things don't remain static. Rather, they *change*. We are now looking at Execution Order and State.

Let's use a few examples to look, slowly and closely, at how *data* change while the code runs.

**Execution order** means code runs line by line, top to bottom (unless a loop or function call redirects it). The sequence matters:

```python
x = 5
x = x + 1
print(x)   # prints 6, not 5
```

**State** is the current value of all variables at any given moment during execution. As the program runs, state changes constantly.

```python
score = 0          # state: score is 0
score = score + 10 # state: score is now 10
score = score + 5  # state: score is now 15
print(score)       # prints 15
```

#### Why This Matters

- Reading code as static text will mislead you
- You need to **mentally trace execution** — imagine running the code step by step in your head
- Ask yourself at each line: *"What are the current values of my variables right now?"*

#### A Practical Technique: The State Table

When reading unfamiliar code, track state manually:

```python
a = 2
b = 3
a = a + b
b = a - b
a = a - b
```

| Line | a | b |
|---|---|---|
| `a = 2` | 2 | — |
| `b = 3` | 2 | 3 |
| `a = a + b` | 5 | 3 |
| `b = a - b` | 5 | 2 |
| `a = a - b` | 3 | 2 |

This is how experienced programmers debug code mentally. It's a skill, and it's learnable.

### A Few More Words On Reading Code

Above we have gone through some basic concepts and building blocks with simple code examples. 

Next we will have a look at a few tips to read Python code clearly. This helps us to develop good habit and learn to write, hopefully, well-structured clean code in the nearest future. 

#### Read Code Structurally

Don't start with details. Read in layers:

1. **Scan the structure** — where are the functions, loops, and conditionals?
2. **Follow the data flow** — what goes in, what comes out, what changes?
3. **Trace execution order** — what runs first? What triggers what?
4. **Read the details** — now zoom into individual lines

Think of it like reading a building blueprint:
- First: how many floors, where are the walls?
- Then: where do the pipes run?
- Finally: what are the light fittings?

#### A Path to Fluency

Learning to read code is like learning a language — with one key difference:

**Language fluency:** exposure + pattern recognition  
**Code fluency:** exposure + pattern recognition + **active execution tracing**

Progress looks like this:

- *Stage 1:* Can't understand individual lines
- *Stage 2:* Understand individual lines, but not the whole
- *Stage 3:* Understand blocks and structure
- *Stage 4:* Can predict what comes next
- *Stage 5:* Can spot bugs before running the code

The gap between Stage 2 and Stage 3 is usually structural reading. The gap between Stage 3 and Stage 4 is execution tracing.

### Practice the Four Building Blocks

These four things cover 80% of beginner code:

- **Variables** — how data is stored and named
- **Functions** — how actions are defined and called
- **Conditionals** — how decisions are made
- **Loops** — how repetition is controlled

Master these, plus the habit of **tracing execution and tracking state**, and you can read most entry-level project code.

Frameworks, classes, decorators, modules — those are later layers. Learn to read everyday conversation before tackling literature.
